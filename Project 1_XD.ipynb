```python
class MAKE_NODE:
    #create node with depth and f(n) value
    def __init__(self,puzzel,depth,f):
        self.puzzel = puzzel
        self.depth = depth
        self.f = f


    #create children nodes (EXPAND)
    def expand(self):
        #firstly let's find the position of zero to move
        for i in range(0, len(self.puzzel)):
            for j in range(0, len(self.puzzel)):
                if self.puzzel[i][j] == '0':
                    a = i
                    b = j
                
        
        #we can move zero to at most four directions, make a list for the candidate movement
        move_candidates = [[a,b+1],[a,b-1],[a+1,b],[a-1,b]]
        children = []

        #switch a,b with move_candidate i[0],i[1]
        for i in move_candidates:
            #make sure the switching is not out of range
            if 0 <= i[0] < len(self.puzzel) and 0 <= i[1] < len(self.puzzel):
                child = []
                for k in self.puzzel:
                    row = []
                    for l in k:
                        row.append(l)
                    child.append(row)
                    
                #print(i[0])
                tmp = child[i[0]][i[1]]
                child[i[0]][i[1]] = child[a][b]
                child[a][b] = tmp
                
                #the node is valid, create node with inclemented depth and default f(n) value
                child_node = MAKE_NODE(child,self.depth+1,0)
                children.append(child_node)
        return children

```

```python
class problem:
    #create puzzel problem by desired dimension
    def __init__(self,dimension):
        self.dimension = dimension
        self.queue = []
        
    
    #define uniforma cost h function (zero)
    def h_uc(self, begin, goal):
        return 0
        
    #define misplaced h function
    def h_misplaced(self, begin, goal):
        score = 0
        for i in range(0, self.dimension):
            for j in range(0, self.dimension):
                if begin[i][j] != '0' and begin[i][j] != goal[i][j]:
                    score += 1
        return score
    
    #define Manhattan distance h function
    def h_manhattan(self, begin, goal):
        array = []
        score = 0
        for i in range(0, self.dimension):
            for j in range(0, self.dimension):
                array.append(begin[i][j])
        
        for i, num in enumerate(array):
            if num != '0':
                score += abs((int(num)-1)%3 - i%3) + abs((int(num)-1)//3 - i//3)
        
        return score
    
    def h(self, begin, goal,alg):
        if alg == 1:
            return self.h_uc(begin, goal)
        if alg == 2:
            return self.h_misplaced(begin, goal)
        if alg == 3:
            return self.h_manhattan(begin, goal)
    
    def play(self):
        goal = [['1','2','3'],['4','5','6'],['7','8','0']]
        
        #input a puzzel to begin
        print("Enter a puzzel to start! (add space in between, use '0' for blank)\n")
        begin = []
        for i in range(0,self.dimension):
            row = input().split(" ")
            begin.append(row)
        
        #create a zero depth zero f score node with begin puzzel
        begin = MAKE_NODE(begin,0,0)
        
        
        #choose algorithm
        print("\nEnter the algorithm you want to use!\n1. Uniform Cost Search.\n2. A* with the Misplaced Tile heuristic. \n3. A* with the Manhattan Distance heuristic.")
        alg = int(input())
        
        
        
        #calculate f score
        if alg in (1,2,3):
            begin.f = self.h(begin.puzzel,goal,alg) + begin.depth
#             else:
#                 print("Invalid input!")
#                 break
        
        #add puzzel node in QUEUE
        self.queue.append(begin)
                   
        space = 0
        while True:
            #if EMPTY(nodes) then return "failure"
            if not self.queue:
                print("Puzzel results in failure!")
                return
            
            cur = self.queue[0]
            
            space += 1
            print("  | ")
            print("  | ")
            print(" \\\'/ \n")
            for i in cur.puzzel:
                for j in i:
                    print(j,end=" ")
                print("")
                
            #if problem.GOAL-TEST(node.STATE) succeeds then return node
            if alg != 1 and (self.h(cur.puzzel,goal,alg)) == 0:
                print("\nPuzzel succeeds in " + str(cur.depth) + " depth and " + str(space) + " nodes!")
                break
            elif alg == 1 and self.h(cur.puzzel, goal,2) == 0:
                print("\nPuzzel succeeds in " + str(cur.depth) + " depth and " + str(space) + " nodes!")
                break
                
            #if not succeeds, refer to QUEUEING-FUNCTION to EXPAND children nodes
            for child in cur.expand():
                child.f = self.h(child.puzzel,goal,alg) + child.depth
                #print(child.f)
                self.queue.append(child)
            del self.queue[0]
            
            #sort nodes in the queue with f score
            self.queue.sort(key = lambda a:a.f,reverse=False)
                

                
        
        
                
        
```
